```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import os, json
import pandas as pd
import numpy as np
import seaborn as sns

def checkData():
    path_to_jsonfiles = "./data"
    cols = ['ranking_type', 'total_xp', 'gender', 'rank_xp', 'rank_random']
    data = pd.DataFrame(columns=cols)
    for file in os.listdir(path_to_jsonfiles):
        full_filename = "%s/%s" % (path_to_jsonfiles, file)
        with open(full_filename,'r') as fi:
            dict = json.load(fi)
            ranking_type = file[len(file)-8:len(file)-5]
            #print(ranking_type)
            candidates = []
            for cand in dict["totaldata"]:
                data = data.append({'ranking_type': ranking_type, 'total_xp' : cand["totalExp"], 'gender' : cand["gender"], 'rank_xp' : cand["xp_ranking"], 'rank_random' : cand["totalRank"]}, ignore_index=True)
    return data
            
checkData()
```

```python
data = checkData()
#plot_data = get_ranked_candidates_experience_gender_rank_task_rankingtype(data)
data = data[(data["ranking_type"].isin(['m_0','f_0']))]
#plot_data = plot_data[(plot_data["task_nr"] == 3)]
data_f = data[(data['gender'] == 'f')]
data_m = data[(data['gender']== 'm')]
s_f = [500] * len(data_f['total_xp'])
s_m = [100] * len(data_m['total_xp'])
fig = plt.figure(figsize=(15, 10))
ax=fig.add_axes([0,0,1,1])
ax.scatter(data_f["rank_xp"], data_f["total_xp"], color='r', s=s_f, marker='+')
ax.scatter(data_m["rank_xp"], data_m["total_xp"], color='b', s=s_m, marker='o')
plt.xticks(np.arange(1, 22, step=1))
plt.yticks(np.arange(0,20, step=1))
ax.set_xlabel('Rank')
ax.set_ylabel('Total Experience')
plt.show()
```
